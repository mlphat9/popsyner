# Global variables

```python
TRAIN_DF_PATH = 'data/train_df.csv'


# ------------- Model Training ------------- #
EPOCHS = 1000
BATCH_SIZE = 100

MODEL_PATH = 'model/ctgan.pkl'
SAVING_PATH = 'results/ctgan.csv'

# ------------- Generating Data ------------- #
GENERATING_SIZE = 7000

print(f'{MODEL_PATH = }')
print(f'{SAVING_PATH = }')
```

# Training phase

## Import Training Set

```python
train_df = pd.read_csv(TRAIN_DF_PATH)
```

```python
train_df.info()
```

```python
train_df.head()
```

## Train model

```python
# Names of the columns that are discrete
discrete_columns = ['sex', 'age2', 'edattain', 'classwk', 'marst', 'geo2_th2000']
```

```python
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=train_df)

synthesizer = CTGANSynthesizer(
    metadata=metadata, # required
    enforce_rounding=False,
    epochs=EPOCHS,
    verbose=True
)

# get the start time
st = time.time()

synthesizer.fit(train_df)

# get the end time
et = time.time()

# get the execution time
elapsed_time = et - st
print(f"Training time: {elapsed_time:.2f} seconds")
```

```python
synthesizer.save(MODEL_PATH)
```

# Generating phase

```python
# Load model
model = CTGANSynthesizer.load(MODEL_PATH)
```

```python
# get the start time
st = time.time()

synthetic_data = model.sample(num_rows=GENERATING_SIZE)

# get the end time
et = time.time()

# get the execution time
elapsed_time = et - st
print(f"Generating time: {elapsed_time:.2f} seconds")
```

```python
synthetic_data.head()
```

```python
# Save the generated data
synthetic_data.to_csv(SAVING_PATH, index=False, header=True)
print(f'{SAVING_PATH = }')
```
